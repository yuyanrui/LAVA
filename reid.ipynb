```python
import json
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
similarity = json.load(open('similarity.json'))
labels = json.load(open('label.json'))
```

```python
len(similarity), len(labels)
similarity = [sim if sim>0 else 0 for sim in similarity]
```

```python
label_pred = [1 if sim>0.8 else 0 for sim in similarity]
fpr, tpr, threshold = roc_curve(label_pred, labels)
roc_auc = auc(fpr, tpr)
roc_auc
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score
precision = precision_score(labels, label_pred)
recall = recall_score(labels, label_pred)
precision, recall
```
